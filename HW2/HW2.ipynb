### 2. Combining results
- `df.groupby(['column_names'])` groups the data by the specified columns.

- The `.size()` function then counts the number of occurrences for each group.

- `reset_index(name='count')` transforms the result into a DataFrame and names the count column as 'count'.

Try to groupby `level` and `is_paid` in udemy dataset and reset_index name by `count`

```python
# Example
species_petal_width = iris.groupby(['petal_width', 'species']).size().reset_index(name='count')
species_petal_width = species_petal_width.sort_values(by='petal_width', ascending=False)
species_petal_width
```

```python
sns.barplot(x='petal_width', y='count', hue='species', data=species_petal_width)
plt.show()
```
